```python
# 6. Write a code to perform basic string compression using the counts of repeated characters.

#CODE

gvnstring = input('Enter the string: ')
compressed_string = ''
count = 1
for i in range(1, len(gvnstring)):
    if gvnstring[i] == gvnstring[i-1]:
        count += 1
    else:
        compressed_string += gvnstring[i-1] + str(count)
        count = 1
compressed_string += gvnstring[-1] + str(count)
print(compressed_string)
```

```python
# 32. Write a code that takes two dictionaries as input and merges them into a single dictionaries. If there are common keys the values should be added together.

#CODE

dic1={}
l1=[]
l2=[]
while True:
    a=input('Enter key for dic1: ')
    b=input('Enter its Value : ')
    l1.append(a)
    l2.append(b)
    choice=int(input('1--->More Data\n2--->Exit\nEnter:'))
    if choice==1:
        pass
    elif choice==2:
        break
    else:
        print('Please enter right choice')
        break
dic1={key:value for key,value in zip(l1,l2)}
dic2={}
l3=[]
l4=[]
while True:
    a=input('Enter key for dic2 : ')
    b=input('Enter its Value : ')
    l3.append(a)
    l4.append(b)
    choice=int(input('1--->More Data\n2--->Exit\nEnter:'))
    if choice==1:
        pass
    elif choice==2:
        break
    else:
        print('Please enter right choice')
        break
dic2={key:value for key,value in zip(l3,l4)}
l5=[]
l6=[]
for i in dic1:
    for j in dic2:
        if i==j:
            l6.append(dic1[i]+dic2[j])
        else:
            l6.append(dic1[i])
        l5.append(i)
        l5.append(j)
        l6.append(dic2[j])
dic3={key:value for key,value in zip (l5,l6)}
print(dic1)
print(dic2)
print(dic3)
```
